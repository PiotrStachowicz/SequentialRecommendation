```python
import matplotlib.pyplot as plt
import pandas as pd
```

## 1. Read loss .csv file

```python
data = pd.read_csv('./loss_categories_2014_adam_L2_lamdas_5.csv', sep='\t')

data.head(5)
```

## 2. Plot each loss category

```python
for col_name in data.columns:
    plt.xlabel('Interaction')
    plt.ylabel(f"{col_name} loss")
    plt.title(f"{col_name}")
    plt.plot(range(len(data)), data[col_name])
    plt.show()
```

## 3. Loss correlation (might as well delete this)

```python
for i in range(len(data.columns) - 1):
    for j in range(i + 1, len(data.columns)):
        col1 = data.columns[j]
        col2 = data.columns[i]

        plt.xlabel(f"{col1} loss")
        plt.ylabel(f"{col2} loss")
        plt.title('Correlation')
        plt.scatter(data[col1], data[col2])
        plt.show()
```
