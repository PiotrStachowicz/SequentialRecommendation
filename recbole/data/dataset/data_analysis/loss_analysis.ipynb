# Loss analysis

```python
import csv
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('./test_categories_loss_2014.csv', sep='\t')

data.head(5)
```

```python
for col_name in data.columns:
    plt.xlabel('Interaction')
    plt.ylabel(f"{col_name} loss")
    plt.title(f"{col_name}")
    plt.plot(range(len(data)), data[col_name])
    plt.show()
```
